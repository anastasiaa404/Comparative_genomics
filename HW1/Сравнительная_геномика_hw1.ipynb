````python
!minimap2 -t 16 -x asm20 -a GCF_000001735.4_TAIR10.1_genomic.fna GCA_036941785.1_ASM3694178v1_genomic.fna | samtools sort -o sorted.bam --write-index -
````

-         ,       BCF.

````python
!bcftools mpileup -Ou -f GCF_000001735.4_TAIR10.1_genomic.fna sorted.bam | bcftools call -mv -Ob -o variants.bcf
````

````python
#  BCF  VCF
!bcftools view variants.bcf -Ov -o variants.vcf
````

````python
#  VCF  BCF
!bcftools index variants.bcf
````

````python
#  VCF   BGZF
!bgzip variants.vcf

#   VCF 
!bcftools index variants.vcf.gz
````
